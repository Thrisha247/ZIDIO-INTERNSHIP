```python
# Import the tools we need
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load the cleaned stock data
df = pd.read_csv("apple_stock.csv")

# Make sure the 'Date' column is actually treated like a date
df['Date'] = pd.to_datetime(df['Date'])

# Make 'Date' the index (so it shows on the left side like row labels)
df.set_index('Date', inplace=True)

# We only need the 'Close' column for prediction
data = df['Close']


# Create the ARIMA model â€” 5,1,0 are basic settings (like "easy mode")
model = ARIMA(data, order=(5, 1, 0))
model_fit = model.fit()

# Ask ARIMA to guess the next 30 days
forecast = model_fit.forecast(steps=30)

# Plot the actual prices and forecast
plt.figure(figsize=(10, 4))
plt.plot(data, label="Real Prices")
plt.plot(forecast.index, forecast, label="Predicted Prices (Next 30 Days)", color='red')
plt.title("Stock Price Forecast with ARIMA ðŸ”®")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Step 1: Load your dataset
df = pd.read_csv("Apple stock data.csv")  # ðŸ‘ˆ Use your actual file name
df['Date'] = pd.to_datetime(df['Date'])   # Convert date column
df.set_index('Date', inplace=True)

# Step 2: Select the closing price (adjust column name if needed)
data = df['Close']  # or replace 'Close' with the exact column name like 'AAPL.Close'

# Step 3: Fit ARIMA model
model = ARIMA(data, order=(5, 1, 0))  # (p=5, d=1, q=0) â€” you can change this later
model_fit = model.fit()

# Step 4: Forecast the next 30 days
forecast = model_fit.forecast(steps=30)

# Step 5: Plot original + forecast
plt.figure(figsize=(12,6))
plt.plot(data, label='Historical Close Price')
plt.plot(pd.date_range(data.index[-1], periods=30, freq='D'), forecast, label='Forecasted Price', color='red')
plt.title('Apple Stock Price Forecast (Next 30 Days)')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 1. Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# 2. Download stock data
data = yf.download('AAPL', start='2022-01-01', end='2024-07-01')
data = data.reset_index()

# 3. Prepare data for Prophet
df = data[['Date', 'Close']]
df = df.rename(columns={"Date": "ds", "Close": "y"})

# 4. Create and train the Prophet model
model = Prophet()
model.fit(df)

# 5. Create a future dataframe (for next 30 days)
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# 6. Plot forecast
model.plot(forecast)
plt.title("ðŸ“ˆ Stock Price Forecast using Prophet")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid()
plt.show()

# 7. Plot trend and seasonality components
model.plot_components(forecast)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
data = yf.download("AAPL", start="2022-01-01", end="2023-12-31")
close_prices = data['Close']
model = ARIMA(close_prices, order=(5, 1, 0))
model_fit = model.fit()
forecast = model_fit.forecast(steps=30)
plt.figure(figsize=(12, 6))
plt.plot(close_prices, label='Original')
plt.plot(pd.date_range(start=close_prices.index[-1], periods=30, freq='D'), forecast, color='red', label='Forecast')
plt.title("Apple Stock Price Forecast (ARIMA)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
